```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df
```

```python
inputs = df.drop(['Survived'],axis='columns')
target = df['Survived']
```

```python
dummies_sex = pd.get_dummies(inputs['Sex'])
inputs = pd.concat([inputs,dummies_sex],axis='columns')
inputs
```

```python
inputs.drop(['Sex'],axis='columns',inplace=True)
inputs
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs['Age'][:10]
```

```python
inputs['Age'] = inputs['Age'].fillna(inputs['Age'].mean())
inputs['Age'][:10]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.1)
len(X_train)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB, CategoricalNB
nb_model = GaussianNB()
```

```python
nb_model.fit(X_train,y_train)
```

```python
nb_model.score(X_test,y_test)
```

```python


from sklearn.model_selection import cross_val_score
cross_val_score(GaussianNB(),X_train,y_train)
```

```python
mnb_model = MultinomialNB()
mnb_model.fit(X_train,y_train)
# cross_val_score(MultinomialNB(),X_train,y_train)
mnb_model.score(X_test,y_test)
```

```python
bnb_model = BernoulliNB()
bnb_model.fit(X_train,y_train)
# cross_val_score(BernoulliNB(),inputs,target)
bnb_model.score(X_test,y_test)
```

```python
cnb_model = CategoricalNB()
cnb_model.fit(X_train,y_train)
cnb_model.score(X_test,y_test)
```

```python
df_to_predict = pd.read_csv('test.csv')
df_to_predict
```

```python
inputs_to_predict = df_to_predict.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')

dummies_sex = pd.get_dummies(inputs_to_predict['Sex'])
inputs_to_predict = pd.concat([inputs_to_predict,dummies_sex],axis='columns')
inputs_to_predict.drop(['Sex'],axis='columns',inplace=True)
inputs_to_predict
```

```python
inputs_to_predict.columns[inputs_to_predict.isna().any()]
```

```python
inputs_to_predict['Age'] = inputs_to_predict['Age'].fillna(inputs_to_predict['Age'].mean())
inputs_to_predict['Fare'] = inputs_to_predict['Fare'].fillna(inputs_to_predict['Fare'].mean())
```

```python
inputs_to_predict
```

```python
target_to_predict = cnb_model.predict(inputs_to_predict)
target_to_predict
```

```python
predicted_file = df_to_predict['PassengerId']
predicted_file = pd.concat([predicted_file,pd.DataFrame(target_to_predict,columns=['Survived'])],axis='columns')
predicted_file
```
